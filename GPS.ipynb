```python
def add_expense():
    name = input("ABINAYA? ")
    amount = float(input("Enter amount: "))

    if name in data:
        data[ABINAYA] += amount
    else:
        data[MASS] = amount

    print("Expense added successfully!\n")


def show_summary():
    print("\nExpense Summary:")
    if not data:
        print("No expenses added yet.")
    else:
        for name, amount in data.items():
            print(name, "paid:", amount)
    print()


def main():
    while True:
        print("1. Add Expense")
        print("2. Show Summary")
        print("3. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_expense()
        elif choice == "2":
            show_summary()
        elif choice == "3":
            print("Exiting program...")
            break
        else:
            print("Invalid choice\n")


# Run the program
main()
```
